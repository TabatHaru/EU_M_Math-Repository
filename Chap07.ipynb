```python
idx = pd.date_range('2015/01/01', '2015/12/31 23:59', freq='T')
# Adjust the size of the random numbers to match the index
dn = np.random.randint(2, size = len(idx)) * 2 - 1

rnd_walk = np.cumprod(np.exp(dn * 0.0002)) * 100

df = pd.Series(rnd_walk, index=idx).resample('B').ohlc()

df.plot(figsize =(15,6),legend= 'best', grid=True)
```

```python
# Comprehensive7-1
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00312/dow_jones_index.zip'

r = requests.get(url, stream=True)

z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
```

```python
dow_jones_index = pd.read_csv('dow_jones_index.data', sep=',')
dow_jones_index.head()
dow_jones_index.info()
```

```python
dow_jones_index_date = pd.to_datetime(dow_jones_index['date'])
delete_dolchar = lambda x: str(x).replace('$', '')

dow_jones_index.open = pd.to_numeric(dow_jones_index.open.apply(delete_dolchar))
dow_jones_index.high = pd.to_numeric(dow_jones_index.high.apply(delete_dolchar))
dow_jones_index.low = pd.to_numeric(dow_jones_index.low.apply(delete_dolchar))
dow_jones_index.close = pd.to_numeric(dow_jones_index.close.apply(delete_dolchar))

```

```python
dow_jones_index.head()
```

```python
dow_jones_index.next_weeks_open = pd.to_numeric(dow_jones_index.next_weeks_open.apply(delete_dolchar))
dow_jones_index.next_weeks_close = pd.to_numeric(dow_jones_index.next_weeks_close.apply(delete_dolchar))
```

```python
dow_jones_index.head()
```

```python
dow_jones_index_stock_index = dow_jones_index.set_index(['date', 'stock'])
dow_jones_index_stock_index_unstack = dow_jones_index_stock_index.unstack()


dow_close_data = dow_jones_index_stock_index_unstack['close']

dow_close_data.describe()
```

```python
corr_data = dow_close_data.corr()
corr_data
```

```python
max_corr = 0
stock_1 = ''
stock_2 = ''


for i in range(0, len(corr_data)):
  print(
      corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[1],
      corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[0],
      corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]][0]

  )

  if max_corr < corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]][0]:
    max_corr = corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]][0]
    stock_1 = corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[1]
    stock_2 = corr_data[i:i+1].unstack().sort_values(ascending=False)[[1]].idxmax()[0]

print('[Max Corr]:', max_corr)
print('[Stock 1]:', stock_1)
print('[Stock 2]:', stock_2)
```

```python
dow_close_data_subsets = dow_close_data[[stock_1, stock_2]]
dow_close_data_subsets.plot(subplots=True, grid=True)
plt.grid(True)
```

```python
log_ratio_stock_close = np.log(dow_close_data/dow_close_data.shift(1))

max_val_stock = log_ratio_stock_close.std().idxmax()
min_val_stock = log_ratio_stock_close.std().idxmin()

print('max volatility :', max_val_stock)
print('min volatility :', min_val_stock)

log_ratio_stock_close[max_val_stock].plot()
log_ratio_stock_close[min_val_stock].plot()
plt.ylabel('log ratio')
plt.legend(loc='best')
plt.grid(True)
```

```python
from jupyter_dash import JupyterDash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# Google Driveをマウント
from google.colab import drive
drive.mount('/content/drive')

# CSVファイルを読み込む
file_path = '/content/drive/My Drive/CHAINAT_pH.csv'
df = pd.read_csv(file_path)

# Dashアプリケーションを作成
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H4('Analysis of Data using scatter matrix'),
    dcc.Dropdown(
        id="dropdown",
        options=[
            {'label': col, 'value': col} for col in df.columns if col != 'species'
        ],
        value=[df.columns[0], df.columns[1]],  # デフォルトの値を設定
        multi=True
    ),
    dcc.Graph(id="graph"),
])

@app.callback(
    Output("graph", "figure"),
    Input("dropdown", "value"))
def update_bar_chart(dims):
    fig = px.scatter_matrix(
        df, dimensions=dims, color="species")
    return fig

# Dashアプリケーションを実行
app.run_server(mode='inline', debug=True)
```
